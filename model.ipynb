```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('fastfood_calories.csv')
pd.set_option('display.max_columns', None)
```

```python
df.dropna(subset=['protein','fiber'],inplace=True)
columns_to_fill = ['vit_a', 'vit_c', 'calcium']

# Fill null values in the specified columns with the mean
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].mean())
```

```python
df.isna().sum()
```

```python
replace = {'Mcdonalds':1, 'Chick Fil-A':2, 'Sonic':3, 'Arbys':4, 'Burger King':5,
       'Dairy Queen':6, 'Subway':7, 'Taco Bell':8}
df['restaurant'] = df['restaurant'].replace(replace)
```

```python
df = df.drop(columns=['Unnamed: 0','salad'])
```

```python
df.info()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

dari data diatas, nilai dengan korelasi tertinggi adalah cal_fat, total_fat, sat_fat, trans_fat, cholesterol, sodium, total_carb, protein

```python
features = ['cal_fat','total_fat','sat_fat','trans_fat','cholesterol','sodium','total_carb','protein']
x = df[features]
y = df['calories']
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=34,test_size=0.1)
y_test.shape
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
```

```python
score = lr.score(x_test,y_test)
print('Akurasi model Regresi Linear', score)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print('R2:', r2_score(y_test,pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test, pred)))
```
